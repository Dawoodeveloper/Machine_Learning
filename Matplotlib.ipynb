```python
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1 *np.random.randn(len(x))
```

```python
plt.plot(x,y,lw = 5)
```

```python
plt.plot(x,y,'--')
```

```python
plt.plot(x,y,'o--',ms = 6) # ms is microsize size of points
```

```python
plt.plot(x,y,':',lw = 3)
```

```python
plt.plot(x,y,'1')
```

```python
plt.plot(x,y,'')
```

```python
plt.figure(figsize = (15,4))
plt.plot(x,y,'+',color = 'green')
```

```python
plt.figure(figsize = (7,7))
plt.plot(x,y,'o--',color = 'green',lw = 1, ms = 5, label = "Component 1")
plt.xlabel("Time (s)",fontsize = 20)
plt.ylabel("Velocity (m/s)",fontsize = 20)
plt.legend(loc = "lower right")
```

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize = (7,7))
plt.plot(x,y,'o--',color = 'green',lw = 1, ms = 5, label = "Component 1 x,y")
plt.plot(x2,y2,'',color = 'red',lw = 2, label = "Component 2 x2, y2")
plt.xlabel("Time (s)",fontsize = 20)
plt.ylabel("Velocity (m/s)",fontsize = 20)
plt.legend(loc = "lower right")
```

```python
plt.figure(figsize = (7,4))
plt.plot(x,y,'o',color = 'green',lw = 1, ms = 5, label = "Data")
plt.plot(x2,y2,'',color = 'red',lw = 2, label = "Theory")
plt.ylim(top = 2)
plt.title('Relation between Velocity & Time')
plt.xlabel("Time (s)",fontsize = 10)
plt.ylabel("Velocity (m/s)",fontsize = 10)
plt.legend(loc = "upper right",ncol = 2)
```

```python
textstr1 = '\n'.join((
    r'$\sigma_a = %.4f$' % (np.std(res_a)),
    r'$\sigma_a = %.4f$' % (np.std(res_b))))

textstr2 = '\n'.join((
    r'$\sigma_a = %.4f$' % (np.std(res_a2)),
    r'$\sigma_a = %.4f$' % (np.std(res_b2))))
```

```python
textstr1
```

```python
fig, axes = plt.subplots(2,2,figsize = (7,7))
ax = axes[0][1]
ax.hist(res_a,bins = 50, density = True,histtype = "step",label = "One")
ax.hist(res_a2,bins = 50, density = True,histtype = "step",label = "Two")
ax.legend(loc ="upper right", fancybox = False,facecolor = "pink",edgecolor = "Red")
ax.text(0.06,0.8,textstr1, transform = ax.transAxes, bbox = dict(facecolor = "yellow", edgecolor = "Green"))
fig.text(0.5,-0.04,'$\Delta E$', ha = "center", fontsize = 20)
plt.show()
```

```python
a = np.linspace(-1,1,1000)
x,y = np.meshgrid(a,a)
z = x**2 + x*y
```

```python
plt.contourf(x,y,z)
plt.colorbar()
plt.show()
```

```python
plt.contourf(x,y,z,levels = 20,vmin = 0.0, cmap = "plasma")
plt.colorbar(label = "Color Bar")
plt.show()
```
